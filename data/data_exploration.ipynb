```python
import pandas as pd
import folium
import numpy as np
```

```python
PATH = r"../data/1-6m-accidents-traffic-flow-over-16-years/accidents_2009_to_2011.csv"
df_raw = pd.read_csv(PATH)
df_raw.head()
```

```python
from datetime import datetime

col = ['Date','Time','Longitude','Latitude', 'Accident_Severity']
df_filt = df_raw[col].dropna()
to_datetime_func = lambda s: datetime.strptime(s, '%d/%m/%Y %H:%M')
time_datetime = df_filt.apply(lambda row: to_datetime_func(row['Date'] +' ' + row['Time']), axis=1)
time_index = pd.Index(time_datetime)
```

```python
df = df_filt[['Longitude','Latitude', 'Accident_Severity']].set_index(time_index)
df.iloc[-5:,:]
```

```python
from folium.plugins import HeatMapWithTime,HeatMap
df_sample = df.iloc[:5000,:]
m = folium.Map(
    location=df_sample[['Latitude','Longitude']].mean().values, tiles="Cartodb Positron", zoom_start=11
)

coordinates = df_sample[['Latitude','Longitude']].values
HeatMap(coordinates).add_to(m)
# (data = coordinates, index=time_datetime[:500].astype(datetime).values).add_to(m)
m
```
